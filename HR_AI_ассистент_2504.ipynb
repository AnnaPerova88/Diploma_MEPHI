```python
# =====================================================
# УСТАНОВКА И ИМПОРТЫ
# =====================================================
!pip install streamlit gigachat -q

import streamlit as st
import requests
import json
from gigachat import GigaChat
import time

# =====================================================
# КЛЮЧ GigaChat (ВАШ)
# =====================================================
GIGACHAT_CREDENTIALS = "MDE5ZGMwOTktYzU4MS03NWY2LTg4YmYtMjk2YmVmMzI1MDQ0OjVlMDY2MmZhLWZjN2UtNGIwZi04YTZjLWQ4NzQwNDZkOWQzYg=="  # ВСТАВЬТЕ СЮДА

# =====================================================
# ФУНКЦИИ ДЛЯ РАБОТЫ С GigaChat
# =====================================================
def get_vacancies_list():
    """Получаем список вакансий от GigaChat"""
    with GigaChat(credentials=GIGACHAT_CREDENTIALS, verify_ssl_certs=False) as giga:
        prompt = """
        Напиши список из 5 актуальных вакансий Data Scientist в крупных IT-компаниях России.
        Формат вывода: просто список в виде:
        1. Название вакансии - Компания (город)
        2. ...
        Без лишнего текста.
        """
        response = giga.chat(prompt)
        return response.choices[0].message.content

def get_interview_questions(vacancy_title):
    """Генерируем вопросы для собеседования под конкретную вакансию"""
    with GigaChat(credentials=GIGACHAT_CREDENTIALS, verify_ssl_certs=False) as giga:
        prompt = f"""
        Ты — HR-специалист, который проводит собеседование на вакансию: {vacancy_title}

        Составь 5 вопросов для проверки кандидата. Вопросы должны покрывать:
        - Технические навыки (Python, SQL, ML)
        - Опыт работы с данными
        - Soft skills (работа в команде, решение проблем)

        Формат вывода:
        Вопрос 1: ...
        Вопрос 2: ...
        ...
        """
        response = giga.chat(prompt)
        return response.choices[0].message.content

def evaluate_answer(question, user_answer):
    """Оцениваем ответ пользователя"""
    with GigaChat(credentials=GIGACHAT_CREDENTIALS, verify_ssl_certs=False) as giga:
        prompt = f"""
        Ты — HR-эксперт. Оцени ответ кандидата на вопрос собеседования.

        Вопрос: {question}
        Ответ кандидата: {user_answer}

        Оцени ответ по шкале 1-10 и дай короткий совет (1-2 предложения).

        Формат ответа:
        Оценка: X/10
        Совет: ...
        """
        response = giga.chat(prompt)
        return response.choices[0].message.content


# =====================================================
# STREAMLIT ИНТЕРФЕЙС
# =====================================================
st.set_page_config(page_title="AI HR-Ассистент", page_icon="🤖", layout="wide")

st.title("🤖 AI HR-Ассистент для подготовки к собеседованию")
st.markdown("---")

# Инициализация состояния сессии
if 'step' not in st.session_state:
    st.session_state.step = "vacancies"
    st.session_state.vacancies = None
    st.session_state.selected_vacancy = None
    st.session_state.questions = None
    st.session_state.current_q = 0
    st.session_state.user_answers = []
    st.session_state.evaluations = []

# =====================================================
# ШАГ 1: ПОКАЗЫВАЕМ СПИСОК ВАКАНСИЙ
# =====================================================
if st.session_state.step == "vacancies":
    st.header("📋 Актуальные вакансии Data Scientist")

    with st.spinner("Загрузка вакансий из GigaChat..."):
        if st.session_state.vacancies is None:
            st.session_state.vacancies = get_vacancies_list()

    st.markdown(st.session_state.vacancies)

    st.markdown("---")
    st.subheader("🎯 Выберите вакансию для подготовки:")

    # Парсим список вакансий (простой вариант — ввод вручную)
    selected = st.text_area("Вставьте название интересующей вакансии:",
                            placeholder="Пример: Data Scientist (Senior) - Яндекс (Москва)")

    if st.button("Начать подготовку 🚀"):
        if selected:
            st.session_state.selected_vacancy = selected
            st.session_state.step = "questions"
            st.rerun()
        else:
            st.error("Пожалуйста, введите название вакансии")

# =====================================================
# ШАГ 2: ГЕНЕРИРУЕМ ВОПРОСЫ
# =====================================================
elif st.session_state.step == "questions":
    st.header(f"🎤 Собеседование на позицию:")
    st.info(st.session_state.selected_vacancy)

    if st.session_state.questions is None:
        with st.spinner("Генерация вопросов для собеседования..."):
            st.session_state.questions = get_interview_questions(st.session_state.selected_vacancy)

    # Парсим вопросы из текста
    questions_list = []
    for line in st.session_state.questions.split('\n'):
        if line.strip() and line[0].isdigit():
            questions_list.append(line.split('.', 1)[1].strip())

    st.markdown("---")
    st.subheader(f"❓ Вопрос {st.session_state.current_q + 1} из {len(questions_list)}")

    # Текущий вопрос
    current_question = questions_list[st.session_state.current_q]
    st.markdown(f"**{current_question}**")

    # Поле для ответа
    user_answer = st.text_area("Ваш ответ:", height=150, key=f"answer_{st.session_state.current_q}")

    col1, col2 = st.columns(2)

    with col1:
        if st.button("✅ Отправить ответ"):
            if user_answer.strip():
                # Сохраняем ответ
                st.session_state.user_answers.append(user_answer)

                # Оцениваем ответ
                with st.spinner("Оценка ответа..."):
                    evaluation = evaluate_answer(current_question, user_answer)
                    st.session_state.evaluations.append(evaluation)

                # Переходим к следующему вопросу
                if st.session_state.current_q + 1 < len(questions_list):
                    st.session_state.current_q += 1
                    st.rerun()
                else:
                    st.session_state.step = "results"
                    st.rerun()
            else:
                st.warning("Пожалуйста, напишите ответ")

    with col2:
        if st.button("🔄 Пропустить вопрос"):
            st.session_state.user_answers.append("[Пропущено]")
            st.session_state.evaluations.append("Оценка: Нет ответа")
            if st.session_state.current_q + 1 < len(questions_list):
                st.session_state.current_q += 1
                st.rerun()
            else:
                st.session_state.step = "results"
                st.rerun()

# =====================================================
# ШАГ 3: РЕЗУЛЬТАТЫ
# =====================================================
elif st.session_state.step == "results":
    st.header("📊 Результаты собеседования")

    # Парсим вопросы
    questions_list = []
    for line in st.session_state.questions.split('\n'):
        if line.strip() and line[0].isdigit():
            questions_list.append(line.split('.', 1)[1].strip())

    # Показываем результаты по каждому вопросу
    for i, (q, a, e) in enumerate(zip(questions_list, st.session_state.user_answers, st.session_state.evaluations)):
        with st.expander(f"Вопрос {i+1}: {q[:80]}..."):
            st.markdown(f"**Ваш ответ:** {a}")
            st.markdown(f"**Оценка HR:** {e}")

    st.markdown("---")

    # Итоговый отзыв от GigaChat
    if st.button("📝 Получить итоговый отзыв"):
        with st.spinner("Генерация итогового отзыва..."):
            summary_prompt = f"""
            Кандидат прошел собеседование на позицию: {st.session_state.selected_vacancy}

            Вопросы и ответы:
            {st.session_state.questions}

            Ответы кандидата:
            {st.session_state.user_answers}

            Напиши короткий итоговый отзыв (3-4 предложения): сильные стороны кандидата, зоны роста, рекомендация.
            """
            with GigaChat(credentials=GIGACHAT_CREDENTIALS, verify_ssl_certs=False) as giga:
                response = giga.chat(summary_prompt)
                st.success(response.choices[0].message.content)

    st.markdown("---")
    if st.button("🔄 Начать заново"):
        for key in ['step', 'vacancies', 'selected_vacancy', 'questions', 'current_q', 'user_answers', 'evaluations']:
            if key in st.session_state:
                del st.session_state[key]
        st.rerun()


# =====================================================
# ЗАПУСК (в Colab)
# =====================================================
# Для запуска в Colab используйте:
# !streamlit run /content/whatever/your/file/is.py --server.port 8501 --server.enableCORS false
# Или через ngrok
```

```python
# Простой текстовый HR-чат в Colab
from gigachat import GigaChat

GIGACHAT_CREDENTIALS = "MDE5ZGMwOTktYzU4MS03NWY2LTg4YmYtMjk2YmVmMzI1MDQ0OjVlMDY2MmZhLWZjN2UtNGIwZi04YTZjLWQ4NzQwNDZkOWQzYg=="

print("=" * 50)
print("🤖 HR-Ассистент: подготовка к собеседованию")
print("=" * 50)

with GigaChat(credentials=GIGACHAT_CREDENTIALS, verify_ssl_certs=False) as giga:
    # 1. Показываем вакансии
    vacancies = giga.chat("Напиши 5 вакансий Data Scientist в России (название и компания)")
    print("\n📋 Вакансии:\n", vacancies.choices[0].message.content)

    # 2. Выбираем вакансию
    vacancy = input("\n👉 Введите название вакансии для подготовки: ")

    # 3. Генерируем вопросы
    questions_prompt = f"""
    Ты HR. Составь 5 вопросов для собеседования на вакансию: {vacancy}
    Формат: просто список вопросов.
    """
    questions = giga.chat(questions_prompt)
    questions_text = questions.choices[0].message.content

    print("\n" + "=" * 50)
    print("❓ Вопросы для собеседования:")
    print("=" * 50)
    print(questions_text)

    # 4. Задаем вопросы и оцениваем
    answers = []
    for i, q_line in enumerate(questions_text.split('\n')):
        if q_line.strip() and '?' in q_line:
            print(f"\n{i+1}. {q_line}")
            answer = input("Ваш ответ: ")
            answers.append(f"Вопрос: {q_line}\nОтвет: {answer}")

            # Оценка
            eval_prompt = f"Оцени ответ на вопрос '{q_line}'. Ответ: {answer}. Оценка 1-10 и совет."
            evaluation = giga.chat(eval_prompt)
            print(f"📝 {evaluation.choices[0].message.content}")

    # 5. Итог
    print("\n" + "=" * 50)
    print("📊 ИТОГ СОБЕСЕДОВАНИЯ")
    print("=" * 50)

    summary_prompt = f"""
    Кандидат отвечал на вопросы. Вот диалог:
    {' '.join(answers)}
    Дай итоговую рекомендацию.
    """
    final = giga.chat(summary_prompt)
    print(final.choices[0].message.content)
```
